# Analyzing the cleaned data

```python
hotels.head()
```

```python
cancelled_perc = hotels['is_canceled'].value_counts(normalize=True)

cancelled_perc.head()
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

ax.bar(
    ["Not canceled", "Canceled"],
    hotels['is_canceled'].value_counts(),
    edgecolor = 'k',
    width = 0.7
)

ax.set_title("Reservation Status Count");
```

```python
hotels.head()
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

sns.countplot(
   x = 'hotel',
   hue = 'is_canceled',
   data = hotels,
   palette = 'Blues'
)

ax.set_title("Reservation Status in Different Category of Hotels");
ax.legend(['Not canceled', 'Canceled']);
ax.set_ylabel("Number of reservations");
```

```python
resort_hotel = hotels[hotels['hotel'] == 'Resort Hotel']
resort_hotel.is_canceled.value_counts(normalize=True)
```

```python
city_hotel = hotels[hotels['hotel'] == 'City Hotel']
city_hotel.is_canceled.value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date').agg({"adr":"mean"})

resort_hotel.head()
```

```python
city_hotel = city_hotel.groupby('reservation_status_date').agg({"adr":"mean"})
city_hotel.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(
    resort_hotel.index,
    resort_hotel['adr'],
    label = "Resort Hotel"
)
ax.plot(
    city_hotel.index,
    city_hotel['adr'],
    label = "City Hotel"
)

ax.set_title("Average Daily Rate in City and Resort Hotel");
ax.legend()
ax.set_ylabel("Average Daily Rate ($)")
```

```python
hotels['month'] = hotels['reservation_status_date'].dt.month
```

```python
fig, ax = plt.subplots(figsize = (16,8))

sns.countplot(
    x = 'month',
    hue = 'is_canceled',
    data = hotels
)

ax.legend(['Not Canceled', 'Canceled'])
ax.set_title("Cancellations Per Month", fontsize = 20)
```

```python
fig, ax = plt.subplots(figsize = (16,8))

sns.barplot(
    x = "month",
    y = "adr",
    data = hotels[hotels['is_canceled'] == 1].groupby('month').agg({"adr":"sum"}).reset_index()
)

ax.set_title("ADR Per Month", fontsize = 20)
```

```python
cancelled_data = hotels[hotels['is_canceled'] == 1]
```

```python
top10_country = cancelled_data['country'].value_counts()[:10]
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.pie(
    top10_country,
    startangle= 90,
    labels=top10_country.index,
    autopct='%.2f',
    rotatelabels=True
);

fig.suptitle("Top 10 countries with reservation cancelled", fontsize=15);
```

```python
cancelled_data_adr = cancelled_data.groupby('reservation_status_date')\
                    .agg({"adr":"mean"}).reset_index().sort_values(by="reservation_status_date")
cancelled_data_adr.head()
```

```python
not_cancelled_data = hotels[hotels['is_canceled'] == 0]
```

```python
not_cancelled_data_adr = not_cancelled_data.groupby('reservation_status_date')\
                        .agg({"adr":"mean"}).reset_index().sort_values(by="reservation_status_date")
not_cancelled_data_adr.head()
```

```python
fig, ax = plt.subplots(figsize=(20,6))

ax.plot(
    not_cancelled_data_adr['reservation_status_date'],
    not_cancelled_data_adr['adr'],
    label = "Not Cancelled"
)
ax.plot(
    cancelled_data_adr['reservation_status_date'],
    cancelled_data_adr['adr'],
    label = "Cancelled"
);
fig.suptitle("Average daily rate", fontsize = 20)
ax.set_xlabel("Reservation date", fontsize=15)
ax.set_ylabel("average daily rate($)", fontsize=15)

ax.legend()
```
